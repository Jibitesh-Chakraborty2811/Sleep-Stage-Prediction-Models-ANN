```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import mne
```

```python
raw = mne.io.read_raw_edf('SC4001E0-PSG (2).edf')
annot = mne.read_annotations('SC4001EC-Hypnogram.edf')
raw.set_annotations(annot,emit_warning=False)
```

```python
raw.plot(start=0,duration=30)
```

```python
data = raw.get_data()
df = pd.DataFrame(data)
df = df.transpose()
df
```

```python
annot_df = annot.to_data_frame()
annot_df
```

```python
label = {'Sleep stage 1':1,'Sleep stage 2':2,'Sleep stage 3':3,'Sleep stage W':0,'Sleep stage 4':3,'Sleep stage R':4,'Sleep stage ?':-1}
annot_df['description'] = annot_df['description'].replace(label)
annot_df
```

```python
annot_df.groupby(['description']).sum()
```

```python
annot_list = []
for i in range(0,153):
    dur = int(annot_df.iloc[i,1])
    des = int(annot_df.iloc[i,2])
    for j in range(0,dur):
        annot_list.append(des)
```

```python
cols = ['Stages']
annot_list = pd.DataFrame(annot_list,columns=cols)
annot_list
```

```python
df = df.iloc[0:79500]
df
```

```python
df_stats = df.describe()
```

```python
df_stats = df_stats.transpose()
```

```python
df_stats
```

```python
def norm(x):
    return (x - df_stats['mean']) / df_stats['std']

df = norm(df)
```

```python
df
```

```python
X = df.iloc[20000:70000].values
Y = annot_list.iloc[20000:70000].values
print(X.shape)
print(Y.shape)
```

```python
from tensorflow.keras.utils import normalize,to_categorical
```

```python
Y = to_categorical(Y)
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
X_train.shape[1]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(28, input_shape = (X_train.shape[1],),activation='relu'))        
model.add(Dense(5, activation='softmax'))    

model.compile(loss='categorical_crossentropy',optimizer='rmsprop',metrics=['acc'])
```

```python
model.fit(X_train,Y_train,validation_data=(X_test,Y_test),epochs=100)
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

```python
predictions = np.round(model.predict(X_test))
```

```python
predictions
```

```python
Y_test
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,predictions))
```

```python
model.save('Sleep6.h5')
```
